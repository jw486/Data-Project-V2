# Calendar Year Analysis 1970's - Current

In this section, we will find trends in the calendar years of the data "species-listings-by=years-totals-report.csv"

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
endangered_animals = pd.read_csv("species-listings-by-year-totals-report.csv")
endangered_animals = endangered_animals.sort_values(by="Calendar Year", ascending=True)
endangered_animals.head(5)
```

- Calendar Year = The year of the species listing
- Number of Species Listings = The number of species including plants and animals.

```python
plt.plot(endangered_animals["Calendar Year"], endangered_animals["Number of Species Listings"])

plt.xlabel("Calendar Year")
plt.ylabel("Number of Endangered Species Listings")
plt.title("Endangered Species Listings by Year") #Plotting Calendar Year VS number of species.
```

Insights
- The number of species listed increase rapidly from 1970 to mid 1990's which could be caused due to the Endangered Species Act taking place.
- There are certain periods where there are spikes. This could mean that the species endangered rates fluctuate alot

Specific Timeframe 2023 - 2013

```python
endangered_animals = pd.read_csv("species-listings-by-year-totals-report.csv")
endangered_animals = endangered_animals.sort_values(by="Calendar Year", ascending=True)
endangered_animals.head(5)
timeframe = endangered_animals[(endangered_animals['Calendar Year'] >= 2013) & (endangered_animals['Calendar Year'] <= 2023)]
timeframe #Specific years 2013 - 2023
```

```python
plt.plot(timeframe["Calendar Year"], timeframe["Number of Species Listings"])

plt.xlabel("Calendar Year")
plt.ylabel("Number of Endangered Species Listings")
plt.title("Endangered Species Listings by Year") #Plotting Calendar Year VS number of species 2023-2013
```

Insights 2 
- In the graph we can see that there is a steep increase in 2013 and in 2016.
- The number of endangered species is low during 2017 - 2022 and then it goes up. It is important to remember that covid played a huge role in the years 2019- 2023.
- The reason why some years had very few endangered species was because covid happened and people stopped going outdoors for a period of time which could mean that there could be some missing data on species.

```python
timeframe_average = sum(list(timeframe['Number of Species Listings'])) / timeframe.shape[0] #average number of species per year (2023-2013)
all_years_average = sum(list(endangered_animals['Number of Species Listings'])) / endangered_animals.shape[0] #Average number of species per year ALL YEARS
timeframe_average,all_years_average
```

We can see that the averages are close to eachother which means that the 2023 -2013 decade has alot of species.
