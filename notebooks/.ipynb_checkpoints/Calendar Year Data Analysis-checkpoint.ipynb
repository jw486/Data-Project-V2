# Calendar Year Analysis 1970's - Current

In this section, our group will find trends in the calendar years of the data "species-listings-by=years-totals-report.csv" from the U.S fish and Wildlife Services dataset.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
species_year = pd.read_csv("species-listings-by-year-totals-report.csv")
species_year = species_year.sort_values(by="Calendar Year", ascending=True)
species_year.head(5)
```

 ## 1. Calendar Timeline

```python
sns.lineplot(data = species_year, x = 'Calendar Year',y='Number of Species Listings')
plt.xlabel("Calendar Year")
plt.ylabel("Number of Species Listings")
plt.title("Total Number of Species Listings by Year") #Plotting Calendar Year VS number of species.
```

<div class="alert alert-block alert-info">

- <strong>There are certain periods where there are spikes</strong>. This could mean that the species endangered rates fluctuate alot <br>
- <strong>many periods with low/baseline activity</strong> (1975,early 1980's,2000-2010,and 2020) in the data that could imply that  there is missing data or there was a higher conservation success rate which decreased the amount of species.

# Summary Observations

<div class="alert alert-block alert-warning">
By looking at the U.S Fish and Wildlife services calendar year species dataset, some major observations can be made. Analyzing the full timeline reveals that the species listing from the 1970's to the end of the 20th century was at its highest and most consistant which was most likely influenced by the Endangered Species Act. Following into the early 21st century we can see that the data becomes very irregular with some spikes of listings while many years had low species listings. This brings us to the focus time period, 2013-2023. In these years we can assume that the ESA enforcement was low with many years having almost no species listed. We found that the average number of species in that decade was lower than the overall average which could be linked to low funding, covid. Covid had a big impact on the data because there reduced field surveys, less funding, and low participation in processing data. Overall, the data suggests that the irregular pattern of species were not entirely due to biodiversity loss but the data gathering itself.
</div>
