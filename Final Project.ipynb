The Endangered Species Act (ESA), established in 1973, provides a legal framework for protecting species at risk of extinction. The central force driving the endangered species act is the U.S Fish and Wildlife Service (FWS) which is the only federal agency dedicated to conservation, protection, and the enhancement of all wildlife in the United States. Under ESA the FWS and the National Marine Fisheries Service (NMFS) collaborate to determine the classification of a targeted species and label them as endangered or threatened. This classification dictates the level of federal protection the species receives. Based on the level of protection, a species is provided with habitat preservation, conservation efforts and even population restoration plans. With the support of government funds and local donations, the FWS provides public access to all their data listings on species recorded over time since the Endangered Species Act was authorized. The FWS is a federal organization dedicated and determined to preserve the beauty of nature.

Understanding when and why certain species face greater risks enables us to refine strategies for protecting biodiversity.

# 2013-2023 Decade Analysis


<div class="alert alert-block alert-warning">
In this notebook, we will be loooking at data containing species of the 6th mass extinction from 2013 through 2023. <br><br>
<b>Explanation:</b> We chose to do 2013-2023 because data from 2025 and 2024 could be incomplete so relying on earlier data makes the data more accurate for interpretation.
</div>

```python
# importing years
data_2023 = pd.read_csv('CSVS/2023.csv')
data_2022 = pd.read_csv('CSVS/2022.csv')
data_2021 = pd.read_csv('CSVS/2021.csv')
data_2020 = pd.read_csv('CSVS/2020.csv')
data_2019 = pd.read_csv('CSVS/2019.csv')
data_2018 = pd.read_csv('CSVS/2018.csv')
data_2017 = pd.read_csv('CSVS/2017.csv')
data_2016 = pd.read_csv('CSVS/2016.csv')
data_2015 = pd.read_csv('CSVS/2015.csv')
data_2014 = pd.read_csv('CSVS/2014.csv')
data_2013 = pd.read_csv('CSVS/2013.csv')
```

```python
# clean function for table 

def clean(table):
    table['Group'] = table['Group'].str.replace(r'<.*?>', '', regex=True) #removing <div> from group column cells
    table["Initial Listing Date"] = pd.to_datetime(table["Initial Listing Date"]) #converting Intial listing Date to a standard time
    table["Year"] = table["Initial Listing Date"].dt.year
    table.rename(columns={"Current\nESA Listing Status  ": "Current ESA Listing Status"}, inplace=True) #rename
    cleaned = table
    
    return cleaned
```

```python
# APPLY CLEAN FUNCTION
df_2023 = clean(data_2023)
df_2022 = clean(data_2022)
df_2021 = clean(data_2021)
df_2020 = clean(data_2020)
df_2019 = clean(data_2019)
df_2018 = clean(data_2018)
df_2017 = clean(data_2017)
df_2016 = clean(data_2016)
df_2015 = clean(data_2015)
df_2014 = clean(data_2014)
df_2013 = clean(data_2013)
```

<div class="alert alert-block alert-warning">
MERGE DATA
</div>

```python
df_list = [df_2023,df_2022,df_2021,df_2020,df_2019,df_2018,df_2017,df_2016,df_2015,df_2014,df_2013] #put tables into list
df_joined = pd.concat(df_list, axis=0, ignore_index=True) #Concatonate the tables to make one big one

df_joined
```

## Graphs

```python


colors = {'Endangered': 'red', 'Threatened': 'orange', 'Resolved Taxon': 'green'} #set colors for ESA Listing Status

status_counts = new_data.groupby(['Locations', 'ESA Listing Status']).size().unstack().fillna(0) #counts for ESA Listing Status

status_counts.plot(kind='bar', stacked=True, figsize=(10, 6), color=[colors[col] for col in status_counts.columns])
plt.xlabel("Region")
plt.ylabel("Count of Animals")
plt.title("ESA Listing Status of Animals by Region")
plt.legend(title="ESA Listing Status", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```

<div class="alert alert-block alert-info">

We can see in the graph, the Pacific Region shows a significantly higher number of endangered animals compared to other regions, suggesting either a greater biodiversity at risk or more active conservation listing/reporting.
</div>

```python
filtered_data = new_data[new_data['ESA Listing Status'].isin(['Endangered', 'Threatened'])]

yearly_animal_counts = filtered_data.groupby(['Locations', 'Type of Animal']).size().unstack().fillna(0)

fig, ax = plt.subplots(figsize=(10, 6))

yearly_animal_counts.plot(kind='bar', ax=ax)

plt.xlabel("Region")
plt.ylabel("Number of Endangered and Threatened Animals")
plt.title("Distribution of Endangered and Threatened Species by Region(2013-2023)")
plt.legend(title="Type of Animal", bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```
